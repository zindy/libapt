```python
from hexdump import hexdump
from pcapfile import savefile
```

```python
testcap = open('test.pcap', 'rb')

capfile = savefile.load_savefile(testcap, verbose=True)

print capfile
```

```python
s = ""
need_concat = False
for p in capfile.packets:
    r = p.raw()
    if len(r) < 31:
        continue

    #string of interest starts at different points for read and write?
    if ord(r[16]) == 0:
        h = 27
    else:
        h = 29
    
    if need_concat:
        s = s + r[h:]
    else:
        s = r[h:]
        
    if ord(r[-1]) == 0x81:
        #Thorlabs says there's another packet coming next
        need_concat = True
    else:
        need_concat = False
        #Apparently r[16] is the direction?!
        if ord(r[16]) == 0: # and ord(r[-2]) == 0x50:
            direction = "sending"
        elif ord(r[16]) == 1:
            direction = "receiving"
        else:
            direction = None
            
        if not direction is None:
            print "Computer is %s: 0x%02x%02x"%(direction, ord(s[1]), ord(s[0]))
            hexdump(s)
```
